```python
class TreeNode:
    def __init__(self, key):
        # Constructor to initialize a tree node
        self.key = key  # Value of the node
        self.left = None  # Reference to the left child node
        self.right = None  # Reference to the right child node

class BinaryTree:
    def __init__(self):
        # Constructor to initialize an empty binary tree
        self.root = None  # The root node of the tree

    def insert(self, key):
        # Insert a key into the tree

        if not self.root:
            # If the tree is empty, create a root node
            self.root = TreeNode(key)
            return

        # Use a recursive helper function to insert the key
        self._insert_recursive(self.root, key)

    def _insert_recursive(self, current, key):
        # Recursive helper function for insertion

        if key < current.key:
            # If the key is smaller, go to the left subtree
            if current.left:
                self._insert_recursive(current.left, key)
            else:
                # If the left child is None, create a new node
                current.left = TreeNode(key)
        elif key > current.key:
            # If the key is larger, go to the right subtree
            if current.right:
                self._insert_recursive(current.right, key)
            else:
                # If the right child is None, create a new node
                current.right = TreeNode(key)
        else:
            # If the key is equal to the current node's key, do nothing (no duplicates allowed)
            return

    def inorder_traversal(self):
        # Perform in-order traversal and return the result as a list

        result = []
        self._inorder_traversal_recursive(self.root, result)
        return result

    def _inorder_traversal_recursive(self, current, result):
        # Recursive helper function for in-order traversal

        if current:
            # Traverse the left subtree
            self._inorder_traversal_recursive(current.left, result)
            # Process the current node
            result.append(current.key)
            # Traverse the right subtree
            self._inorder_traversal_recursive(current.right, result)

# Example usage:
tree = BinaryTree()
keys = [5, 3, 7, 2, 4, 6, 8]

for key in keys:
    tree.insert(key)

# In-order traversal
result = tree.inorder_traversal()
print("In-order Traversal:", result)
```
